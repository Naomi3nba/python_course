```python
#Importing Libreries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:,'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```

```python
# Select First 5 Rows with .iloc()
# Select and display the job_title and job_location columns for the first 5 rows of the DataFrame using the iloc method.

x1 = df.iloc[:5,[0,2]]
print(x1)
```

```python
# Get Rows and Columns with .loc()
# Retrieve the columns starting with job_title all the way to job_posted_date for rows 
# with an index of 5 to 9. Use the loc method.

x2 = df.loc[5:9,['job_title','job_posted_date']]
print(x2)
```

```python
# Retrieve the job_title and salary_year_avg for the first 15 rows 
# where the job_country is 'United States'. Use the iloc method.
x_usa= df[df['job_country'] =='United States']
x3 = x_usa.iloc[:16,[0,13]]

print(x3)
```

```python
# Select and display the job_title, job_location, 
# and job_posted_date columns for rows where the job_posted_date 
# is after '2023-05-01' using the loc method.

x_date1 = df[df['job_posted_date'] > '2023-05-01' ]
x4 = x_date1.loc[:5,['job_title', 'job_location','job_posted_date']]
print(x4)
```
