```python
#Importing Libreries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
median_y = df['salary_year_avg'].median()
print(median_y)
median_h = df['salary_hour_avg'].median()
print(median_h)
```

```python
# DATA CLEANING 

# 1) REPLACE VALUES NaN: 
df_filler = df
df_filler['salary_year_avg'] = df_filler['salary_year_avg'].fillna(median_y) #Fill NaN with other value
df_filler['salary_hour_avg'] = df_filler['salary_hour_avg'].fillna(median_h) #Fill NaN with other value

# 2) Eliminate duplicates values: 
df_unique = df_filler
df_unique = df_unique.drop_duplicates()

df_unique = df_filler
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])
```
